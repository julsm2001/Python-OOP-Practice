### **Creacion de clase Coche**


```python
class Coche():
    def __init__(self, marca: str, modelo: int, kilometros_recorridos: float, gasolina: float):
        self.marca = marca
        self.modelo = modelo
        self.kilometros_recorridos = kilometros_recorridos
        self.gasolina = gasolina

    def avanzar(self, kilometros):
        consumo = kilometros * 0.04

        if self.gasolina >= consumo:
            self.kilometros_recorridos = kilometros + self.kilometros_recorridos
            self.gasolina = self.gasolina - consumo
        else:
            print("Es necesario tanquear para recorrer la cantidad indicada de kilómetros")
    
    def tanquear(self, galones_cargados):
        self.gasolina = self.gasolina + galones_cargados
```

### **Creacion de Objeto**

**a)** Al inicio: kilometros_recorridos = 0; gasolina = 12

```python
mi_coche = Coche("Toyota", 2026, 0.0, 12.0)
```

### **Comprovacion**

**b) avanzar(100):** kilometros_recorridos = 100, gasolina = 8

```python
mi_coche.avanzar(100)

print(f"Km recorridos: {mi_coche.kilometros_recorridos}")
print(f"Gasolina restante: {mi_coche.gasolina}")
```

**c) avanzar(140):** kilometros_recorridos = 240, gasolina = 2,4

```python
mi_coche.avanzar(140)

print(f"Km recorridos: {mi_coche.kilometros_recorridos}")
print(f"Gasolina restante: {mi_coche.gasolina}")
```

**d) avanzar(70):** "Es necesario tanquear para recorrer la cantidad indicada de kilómetros”

```python
mi_coche.avanzar(70)

print(f"Km recorridos: {mi_coche.kilometros_recorridos}")
print(f"Gasolina restante: {mi_coche.gasolina}")
```

**e) tanquear(5):** gasolina = 7,4


```python
mi_coche.tanquear(5)

print(f"Gasolina tras tanquear: {mi_coche.gasolina}")
```

**f) avanzar(70):** kilometros_recorridos = 310, gasolina = 4,6


```python
mi_coche.avanzar(70)

print(f"Km recorridos: {mi_coche.kilometros_recorridos}")
print(f"Gasolina restante: {mi_coche.gasolina}")
```
